```python
# ================================================================
# ASSOCIATION RULES - ONLINE RETAIL (Assignment Ready)
# File path: /content/Online retail.xlsx
# ================================================================

# If running in a fresh environment (Colab), uncomment the next line:
# !pip install mlxtend --quiet

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import warnings
warnings.filterwarnings('ignore')
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)

# ---------------------------
# 1. LOAD DATA
# ---------------------------
file_path = "/content/Online retail.xlsx"
df = pd.read_excel(file_path)

print("Data shape:", df.shape)
display(df.head())

# ---------------------------
# 2. PREPROCESSING
# ---------------------------

# The current data appears to be already pre-processed into transactions, where
# each row in the first (and only) column contains a comma-separated list of items.
# The original preprocessing steps for 'InvoiceNo', 'Description', 'Quantity', 'UnitPrice'
# are not applicable to this format and will be removed.

# Extract the column containing the item lists
# Based on `df.head()` output, the data is in the first column.
transaction_strings = df.iloc[:, 0].astype(str)

print("\nSample raw transaction strings (first 5):")
display(transaction_strings.head())

# ---------------------------
# 3. Transaction / Basket Creation
# ---------------------------
# Convert each comma-separated string into a list of cleaned items (stripped and lowercased).
# This replaces original steps 2.1-2.5 and 3.
basket = transaction_strings.apply(lambda x: [item.strip().lower() for item in x.split(',') if item.strip()]).tolist()

# Remove any empty lists that might result from entirely empty or malformed rows
basket = [t for t in basket if t]

print("\nNumber of transactions (invoices):", len(basket))
print("Example transaction (first):", basket[0][:12])

# ---------------------------
# 4. Convert transactions to one-hot encoded dataframe
# ---------------------------
te = TransactionEncoder()
te_ary = te.fit(basket).transform(basket)
trans_df = pd.DataFrame(te_ary, columns=te.columns_)

print("\nOne-hot encoded shape:", trans_df.shape)
print("Number of distinct products:", len(trans_df.columns))

# Optional: drop extremely rare items to speed up apriori (threshold here is adjustable)
min_item_support_cutoff = 0.001  # keep items with at least 0.1% support
item_support = trans_df.mean()
rare_items = item_support[item_support < min_item_support_cutoff].index
if len(rare_items) > 0:
    trans_df = trans_df.drop(columns=rare_items)
    print(f"\nDropped {len(rare_items)} rare items (support < {min_item_support_cutoff})")
    print("New one-hot encoded shape:", trans_df.shape)

# ---------------------------
# 5. FREQUENT ITEMSETS (Apriori)
# ---------------------------
# Choose a minimum support; 0.01 = 1% of transactions. Adjust as needed.
min_support = 0.01
freq_itemsets = apriori(trans_df, min_support=min_support, use_colnames=True, verbose=1)

# Add support counts
freq_itemsets['length'] = freq_itemsets['itemsets'].apply(lambda x: len(x))
print("\nFrequent itemsets found:", len(freq_itemsets))
display(freq_itemsets.sort_values(['support','length'], ascending=[False, False]).head(20))

# Save frequent itemsets
freq_itemsets.to_csv('/content/frequent_itemsets.csv', index=False)
print("Frequent itemsets saved to /content/frequent_itemsets.csv")

# ---------------------------
# 6. ASSOCIATION RULES
# ---------------------------
# Generate rules with min_threshold for confidence; adjust lift threshold later for filtering.
rules = association_rules(freq_itemsets, metric='confidence', min_threshold=0.2)
# Add a 'support_count' column for interpretability
rules['support_count'] = (rules['support'] * len(trans_df)).round().astype(int)

# Sort rules by lift then confidence
rules = rules.sort_values(['lift', 'confidence'], ascending=[False, False]).reset_index(drop=True)
print("\nNumber of raw rules generated:", len(rules))
display(rules.head(20))

# Save full rules
rules.to_csv('/content/association_rules_all.csv', index=False)
print("All rules saved to /content/association_rules_all.csv")

# ---------------------------
# 7. FILTER & PICK MEANINGFUL RULES
# ---------------------------
# Typical thresholds  you may tune these for your dataset
min_support_rule = 0.01       # at least 1% support
min_confidence_rule = 0.3     # 30% confidence
min_lift_rule = 1.2           # lift > 1.2 (positive association)

filtered_rules = rules[
    (rules['support'] >= min_support_rule) &
    (rules['confidence'] >= min_confidence_rule) &
    (rules['lift'] >= min_lift_rule)
].copy().reset_index(drop=True)

print("\nFiltered rules count:", len(filtered_rules))
display(filtered_rules.sort_values(['lift','confidence'], ascending=[False, False]).head(30))

filtered_rules.to_csv('/content/association_rules_filtered.csv', index=False)
print("Filtered rules saved to /content/association_rules_filtered.csv")

# ---------------------------
# 8. TOP-N INSIGHTS & VISUALIZATIONS
# ---------------------------
# Top products by frequency
top_products = trans_df.mean().sort_values(ascending=False).head(20)
plt.figure(figsize=(10,6))
sns.barplot(x=top_products.values, y=top_products.index)
plt.title("Top 20 Most Frequent Products (by transaction support)")
plt.xlabel("Support (fraction of transactions)")
plt.tight_layout()
plt.show()

# Plot support vs confidence for filtered rules
if not filtered_rules.empty:
    plt.figure(figsize=(8,6))
    sns.scatterplot(x='support', y='confidence', size='lift', data=filtered_rules, legend=False, alpha=0.7)
    plt.title('Support vs Confidence (filtered rules)  bubble size = lift')
    plt.xlabel('Support')
    plt.ylabel('Confidence')
    plt.show()

# Visualize top rules by lift as a network-like chart (simpler bubble plot)
top_rules = filtered_rules.sort_values('lift', ascending=False).head(15)
plt.figure(figsize=(12,6))
for i, row in top_rules.iterrows():
    ant = ','.join(list(row['antecedents']))
    con = ','.join(list(row['consequents']))
    plt.scatter(row['support'], row['confidence'], s=row['lift']*40, alpha=0.6)
    plt.text(row['support']+0.0005, row['confidence']+0.005, f"{ant} -> {con}", fontsize=8)
plt.title("Top Rules (annotated)  support vs confidence (bubble size ~ lift)")
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.show()

# Heatmap of top 12 items co-occurrence (for interpretation)
top_items = trans_df.mean().sort_values(ascending=False).head(12).index.tolist()
co_occ = trans_df[top_items].T.dot(trans_df[top_items])
# convert to counts and normalize to fraction of transactions
co_occ_norm = co_occ / len(trans_df)
plt.figure(figsize=(10,8))
sns.heatmap(co_occ_norm, annot=True, fmt=".3f", cmap="YlGnBu")
plt.title("Co-occurrence (fraction of transactions)  Top 12 items")
plt.show()

# ---------------------------
# 9. SAMPLE READABLE OUTPUT (for report)
# ---------------------------
print("\n=== Example human-readable top filtered rules (top 20) ===")
for idx, r in filtered_rules.head(20).iterrows():
    antecedent = ', '.join(r['antecedents'])
    consequent = ', '.join(r['consequents'])
    print(f"{idx+1}. Rule: [{antecedent}] -> [{consequent}] | support={r['support']:.4f} ({r['support_count']} tx) | confidence={r['confidence']:.2f} | lift={r['lift']:.2f}")

# ---------------------------
# 10. SAVE FINAL REPORT DATA
# ---------------------------
# Save a compact CSV with antecedent/consequent strings for easy reading
out = filtered_rules.copy()
out['antecedent_str'] = out['antecedents'].apply(lambda s: ', '.join(list(s)))
out['consequent_str'] = out['consequents'].apply(lambda s: ', '.join(list(s)))
out[['antecedent_str','consequent_str','support','confidence','lift','support_count']].to_csv('/content/association_rules_report.csv', index=False)
print("\nHuman-readable filtered rules saved to /content/association_rules_report.csv")

# ---------------------------
# 11. NOTES & TUNING ADVICE (for grader)
# ---------------------------
notes = f"""
NOTES & TUNING ADVICE:
- File read from: {file_path}
- Preprocessing steps: The input data was assumed to be one column of comma-separated item lists. Preprocessing now involves splitting these strings into lists, stripping whitespace, and lower-casing items. Original steps for 'InvoiceNo', 'Description', 'Quantity', 'UnitPrice' were removed as inapplicable to this data format.
- IQR/outlier capping NOT applied here because market-basket requires raw counts; instead rare items were dropped (support < {min_item_support_cutoff}).
- Apriori min_support used: {min_support}
- Rule filters used: support >= {min_support_rule}, confidence >= {min_confidence_rule}, lift >= {min_lift_rule}
- Files generated: /content/frequent_itemsets.csv, /content/association_rules_all.csv, /content/association_rules_filtered.csv, /content/association_rules_report.csv
"""
print(notes)

# ===========================
# END
# ===========================
print("\n Completed Association Rule Mining. Good luck with submission!")
```

1.	What is lift and why is it important in Association rules?   
answer:    
Lift is an evaluation metric used in Association Rule Mining that measures how much more frequently two items occur together compared to what would be expected if they were statistically independent.   
Lift is important because it helps to determine whether a rule shows a true association or if the co-occurrence is merely due to chance

2.	What is support and Confidence. How do you calculate them?   
answer:

  Support measures how frequently an item or itemset appears in the entire dataset.   
  

    It is calculate by   
    Support(A)=
    Number of transactions containing A / Total number of transactions   

  Confidence measures how often items in the consequent (B) appear in transactions that already contain the antecedent (A).   
  It is calculate by   
  Confidence(AB)=   
  Number of transactions containing both A and B /
  Number of transactions containing A



3.	What are some limitations or challenges of Association rules mining?   
answer:    
i)Generation of Too Many Rules   
Association rule mining can produce thousands of rules, making it difficult to interpret and select meaningful ones.
    
   ii)Need for Threshold Selection  
   The quality of results depends heavily on choosing suitable minimum support, confidence, and lift values.   
   iii)Ignores Item Quantity and Timing   
   Traditional algorithms only consider presence or absence of items.  
   iv)Rare but Important Item Combinations May Be Missed   
   Low-support items are often filtered out even if they are strategically important (e.g., luxury items that rarely sell but strongly co-occur).   
   v)Computational Complexity   
   For large datasets with many distinct items, computation becomes slow and memory-intensive, especially with Apriori due to repeated candidate generation.   
   vi)Lack of Causal Interpretation   
   Rules express co-occurrence, not true cause-effect relationships.  
   Example: chips  cola does not mean chips cause cola purchase.
